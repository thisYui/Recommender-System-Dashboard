# Khái niệm

Recommeder System là một hệ thống có nhiệm vụ đưa ra hay dự đoán mức độ phù hợp giữa người dùng (user) và đối tượng (item). Định nghĩa toán học thuần túy:

$$ f(user, item) -> score $$

Ta lặp lại hành động trên đối với nhiều đối tượng (item) khác nhau và xắp xếp chúng theo thứ tự từ cao đến thấp và lấy $K$ điểm cao nhất gọi là $Top-K$. 2 góc nhìn chính của Recommender system:
- Tìm kiếm đối tượng (item) phù hợp nhát với người dùng (user): Gọi ý phim ảnh, nhạc, sản phẩn,...
- Tìm người dùng (user) phù hợp nhất với đối tượng (item): Gợi ý quảng cáo, Tinder,...

Bản chât hai bài toán là giống nhau đều tìm giá trị cao nhât của $score$ và xếp hạng chúng, nhưng tùy theo bài toán cụ thể và kích thước user và item có thể thay đổi cách tiếp cận ban đầu. Cả hai cách tiếp cận đều giống nhau:

- Tính score giữa user và item
- Sắp xếp theo score
- Lấy Top-K

> Do đó recommender system thực chất là bài toán ranking.

## Dạng dữ liệu của Recommender System

Thông thường, dữ liệu của Recommender System được biểu diễn dưới dạng **user–item interaction matrix**:

$$
R_{u,i}
$$

Trong đó:

- $u$ : user
- $i$ : item
- $R_{u,i}$ : mức độ tương tác giữa user $u$ và item $i$

Ví dụ:

| user | item1 | item2 | item3 |
|------|-------|-------|-------|
| u1   | 5     | ?     | 3     |
| u2   | ?     | 4     | ?     |
| u3   | 2     | ?     | 5     |

Trong bảng này:

- Các ô thể hiện mức độ quan tâm của user với item (rating, click, purchase, ...)
- Các ô dánh dấu `?` là các giá trị chưa biết (missing interaction)
- Ma trận này thường **rất thưa (sparse)** vì mỗi user chỉ tương tác với một số ít item

Mục tiêu của Recommender System là dự đoán các giá trị còn thiếu:

$$
\hat{R}_{u,i} = f(u,i)
$$

### Sparse Matrix Problem

Trong thực tế, ma trận user–item thường rất thưa (sparse):

- số user rất lớn
- số item rất lớn
- mỗi user chỉ tương tác với rất ít item

Ví dụ:

- 1M user
- 100K item
- mỗi user xem ~50 item

Density:

$$
\frac{50}{100000} = 0.05\%
$$

Do đó, việc tính:

$$
f(u,i) \quad \forall i \in I
$$

là rất tốn chi phí tính toán.

---

### Candidate Generation

Thay vì dự đoán trên toàn bộ item, recommender system thường chia thành 2 bước:

1. Candidate generation
2. Ranking

Bước 1: chọn tập item ứng viên nhỏ hơn

$$
C(u) \subset I
$$

Bước 2: chỉ ranking trong tập này

$$
TopK(u) = \arg\max_{i \in C(u)} f(u,i)
$$

Trong đó:

- $|C(u)| \ll |I|$
- ví dụ: 1000 candidate → chọn top 10

---

### Why

Sparse matrix → không thể predict toàn bộ item
→ phải **giới hạn candidate space**
→ recommender system trở thành bài toán **two-stage ranking**


## Thành phần chính của Recommender System

Một Recommender System điển hình gồm các thành phần chính sau:

### 1. Users

Tập người dùng của hệ thống:

$$
U = \{u_1, u_2, ..., u_m\}
$$

Mỗi user có thể có:

- thông tin hồ sơ (age, gender, location, ...)
- lịch sử tương tác
- hành vi duyệt hệ thống
- embedding vector của user

Ví dụ:

- người xem Netflix
- người mua hàng Shopee
- người nghe nhạc Spotify

---

### 2. Items

Tập đối tượng được gợi ý:

$$
I = \{i_1, i_2, ..., i_n\}
$$

Item có thể là:

- phim
- sản phẩm
- bài hát
- video
- bài viết

Mỗi item có thể có:

- metadata (category, tag, genre)
- nội dung (text, image, audio)
- embedding vector

---

### 3. Interaction

Interaction mô tả hành vi giữa user và item:

$$
R_{u,i}
$$

Các dạng interaction phổ biến:

Explicit feedback:
- rating (1–5 sao)
- like / dislike
- review

Implicit feedback:
- click
- view
- watch time
- purchase
- add to cart

Interaction được dùng để:

- xây dựng user-item matrix
- huấn luyện model
- học preference của user

---

### 4. Model dự đoán

Model học hàm scoring:

$$
f(u,i) \rightarrow score
$$

Score thể hiện:

- mức độ user thích item
- xác suất user click
- khả năng user mua

Một số model phổ biến:

- Collaborative filtering
- Matrix factorization
- Neural recommender
- Content-based model
- Hybrid model

Output của model:

$$
\hat{R}_{u,i}
$$

---

### 5. Ranking Top-K

Sau khi có score cho các item:

$$
f(u,i_1), f(u,i_2), ..., f(u,i_n)
$$

Ta sắp xếp giảm dần:

$$
i_{(1)} \succ i_{(2)} \succ ... \succ i_{(n)}
$$

Sau đó lấy K item đầu:

$$
TopK(u) = \{i_{(1)}, ..., i_{(K)}\}
$$

Danh sách này là recommendation cuối cùng.

---

## Pipeline của Recommender System

Pipeline tổng quát:

> Data → Model → Predict score → Rank → Recommend

Chi tiết hơn:

> User + Item + Interaction → Training Data → Train Model → f(user, item) → score → Ranking → Top-K → Recommendation list

Hoặc dạng 2-stage system (thực tế hơn):

> Data → Candidate Generation → Ranking Model → Top-K → Recommendation


# 3 loại Recommender System phổ biến

## 1. Collaborative Filtering [(Wikipiedia)](https://en.wikipedia.org/wiki/Collaborative_filtering)

Collaborative Filtering (Lọc cộng tác) là phương pháp Recommender System dựa trên **hành vi của người dùng** để đưa ra gợi ý, **không cần biết nội dung của item**. 

Ý tưởng chính:

> Những người dùng có hành vi giống nhau trong quá khứ sẽ có xu hướng thích những item giống nhau trong tương lai.

Theo nghĩa hẹp hơn và mới hơn, lọc cộng tác là một phương pháp đưa ra **dự đoán** tự động (lọc) về sở thích của người dùng bằng cách sử dụng thông tin về sở thích hoặc thị hiếu được thu thập từ nhiều **người dùng** (cộng tác). Cách tiếp cận này giả định rằng nếu người A và B có chung quan điểm về một vấn đề, họ có nhiều khả năng đồng ý với nhau về các vấn đề khác hơn so với việc ghép ngẫu nhiên A với một người khác. Ví dụ, một hệ thống lọc cộng tác cho chương trình truyền hình có thể dự đoán chương trình nào người dùng có thể thích dựa trên danh sách hạn chế về thị hiếu (thích hoặc không thích) của người dùng. Những dự đoán này cụ thể cho từng người dùng, nhưng sử dụng thông tin thu thập được từ nhiều người dùng. Điều này khác với cách tiếp cận đơn giản hơn là đưa ra **điểm trung bình** (không cụ thể) cho mỗi mục quan tâm, ví dụ dựa trên số phiếu bầu của nó.

Theo nghĩa tổng quát hơn, lọc cộng tác là quá trình lọc thông tin hoặc mẫu bằng cách sử dụng các kỹ thuật liên quan đến sự hợp tác giữa nhiều tác nhân, quan điểm, nguồn dữ liệu, v.v. Các ứng dụng của lọc cộng tác thường liên quan đến các tập **dữ liệu rất lớn**. Các phương pháp lọc cộng tác đã được áp dụng cho nhiều loại dữ liệu bao gồm: dữ liệu cảm biến và giám sát; dữ liệu tài chính; và dữ liệu người dùng từ thương mại điện tử và các ứng dụng web.

Động lực thúc đẩy phương pháp lọc cộng tác xuất phát từ ý tưởng rằng mọi người thường nhận được những đề xuất tốt nhất từ ​​những người có sở thích tương tự với họ. Một vấn đề then chốt của phương pháp lọc cộng tác là làm thế nào để kết hợp và trọng số hóa sở thích của những người dùng lân cận. Đôi khi, người dùng có thể đánh giá ngay lập tức các mặt hàng được đề xuất.

---

### Phương pháp luận
Các hệ thống lọc cộng tác có nhiều hình thức, nhưng nhiều hệ thống phổ biến có thể được đơn giản hóa thành hai bước:

1. Hãy tìm những người dùng có cùng mô hình đánh giá với người dùng hiện tại (người dùng mà dự đoán hướng đến).
2. Sử dụng xếp hạng từ những người dùng có cùng quan điểm tìm được ở bước 1 để tính toán dự đoán cho người dùng hoạt động.

Điều này thuộc loại lọc cộng tác dựa trên người dùng. Một ứng dụng cụ thể của nó là thuật toán láng giềng gần nhất dựa trên người dùng. Ngoài ra, **phương pháp lọc cộng tác dựa trên mặt hàng** [Wikipiedia](https://en.wikipedia.org/wiki/Item-item_collaborative_filtering) (người dùng đã mua x cũng đã mua y) hoạt động theo cách tập trung vào mặt hàng:

Xây dựng ma trận tương quan giữa các mục để xác định mối quan hệ giữa các cặp mục.
Suy luận sở thích của người dùng hiện tại bằng cách xem xét ma trận và đối chiếu với dữ liệu của người dùng đó.

Một hình thức lọc cộng tác khác có thể dựa trên những quan sát ngầm về hành vi người dùng thông thường (trái ngược với hành vi nhân tạo được áp đặt bởi nhiệm vụ đánh giá). Các hệ thống này quan sát những gì **một người dùng** đã làm cùng với những gì **tất cả người** dùng đã làm (họ đã nghe nhạc gì, họ đã mua những mặt hàng nào) và sử dụng dữ liệu đó để dự đoán hành vi của người dùng trong tương lai, hoặc để dự đoán cách người dùng có thể hành xử nếu có cơ hội. Sau đó, những dự đoán này phải được lọc qua logic nghiệp vụ để xác định xem chúng có thể ảnh hưởng như thế nào đến các hành động của hệ thống nghiệp vụ. Ví dụ, sẽ không hữu ích nếu chào bán cho ai đó một album nhạc cụ thể nếu họ đã chứng minh rằng họ sở hữu album nhạc đó.

---
### Các loại Collaborative Filtering

Collaborative Filtering có thể được chia thành các nhóm chính:

1. Memory-based Collaborative Filtering
2. Model-based Collaborative Filtering
3. Hybrid Collaborative Filtering
4. Deep Learning Collaborative Filtering
5. Context-aware Collaborative Filtering

---

### 1. Memory-based Collaborative Filtering

Memory-based Collaborative Filtering là phương pháp sử dụng trực tiếp **user–item interaction matrix** để đưa ra dự đoán, không cần huấn luyện mô hình.

Ý tưởng chính:

> Nếu hai người dùng có hành vi giống nhau trong quá khứ, họ sẽ có xu hướng thích các item giống nhau trong tương lai.

---

#### User-based Collaborative Filtering

Phương pháp này tìm các user giống nhau, sau đó dự đoán rating dựa trên rating của họ.

Dự đoán rating:

$$
\hat r_{u,i}
=
\sum_{v \in N(u)}
sim(u,v) \cdot r_{v,i}
$$

Trong đó:

- $u$ : user cần dự đoán
- $i$ : item cần dự đoán
- $v$ : user lân cận của $u$
- $N(u)$ : tập user giống với user $u$
- $sim(u,v)$ : độ tương đồng giữa user $u$ và $v$
- $r_{v,i}$ : rating của user $v$ với item $i$
- $\hat r_{u,i}$ : rating dự đoán

Để tránh scale lớn, thường chuẩn hóa:

$$
\hat r_{u,i}
=
\frac{
\sum_{v \in N(u)}
sim(u,v) r_{v,i}
}
{
\sum_{v \in N(u)}
|sim(u,v)|
}
$$

Ý nghĩa:

- tử số: weighted rating từ user giống nhau
- mẫu số: chuẩn hóa similarity

---

#### Item-based Collaborative Filtering

Phương pháp này tìm item giống nhau.

Dự đoán:

$$
\hat r_{u,i}
=
\sum_{j \in I(u)}
sim(i,j) r_{u,j}
$$

Trong đó:

- $u$ : user
- $i$ : item cần dự đoán
- $j$ : item user đã tương tác
- $I(u)$ : tập item user $u$ đã rating
- $sim(i,j)$ : similarity giữa item
- $r_{u,j}$ : rating user u với item j

Chuẩn hóa:

$$
\hat r_{u,i}
=
\frac{
\sum_{j \in I(u)}
sim(i,j) r_{u,j}
}
{
\sum_{j \in I(u)}
|sim(i,j)|
}
$$

---

### Similarity Measures

#### Cosine Similarity

Độ tương đồng giữa hai user:

$$
sim(u,v)
=
\frac{
\sum_i r_{u,i} r_{v,i}
}
{
\sqrt{\sum_i r_{u,i}^2}
\sqrt{\sum_i r_{v,i}^2}
}
$$

Trong đó:

- $r_{u,i}$ : rating user u với item i
- $r_{v,i}$ : rating user v với item i
- tử số: dot product giữa hai vector
- mẫu số: chuẩn hóa độ dài vector

---

#### Pearson Correlation

$$
sim(u,v)
=
\frac{
\sum_i (r_{u,i}-\bar r_u)(r_{v,i}-\bar r_v)
}
{
\sqrt{\sum_i (r_{u,i}-\bar r_u)^2}
\sqrt{\sum_i (r_{v,i}-\bar r_v)^2}
}
$$

Trong đó:

- $\bar r_u$ : rating trung bình của user u
- $\bar r_v$ : rating trung bình của user v
- công thức đo tương quan tuyến tính

---

### 2. Model-based Collaborative Filtering

Model-based Collaborative Filtering học một model dự đoán:

$$
f(u,i) \rightarrow score
$$

thay vì dựa trực tiếp vào similarity.

---

#### Matrix Factorization

Ý tưởng:

Phân rã ma trận user-item thành hai ma trận latent:

$$
R \approx U V^T
$$

Trong đó:

- $R$ : ma trận user-item
- $U$ : user latent matrix
- $V$ : item latent matrix
- $U_u$ : vector latent của user u
- $V_i$ : vector latent của item i

Score:

$$
\hat r_{u,i}
=
U_u \cdot V_i
$$

Trong đó:

- $U_u$ : vector latent user
- $V_i$ : vector latent item
- dot product thể hiện similarity trong latent space

---

#### Optimization Objective

Model được học bằng cách minimize:

$$
\min
\sum_{(u,i) \in R}
(r_{u,i} - U_u V_i)^2
+
\lambda
(
||U_u||^2 + ||V_i||^2
)
$$

Trong đó:

- $r_{u,i}$ : rating thật
- $\hat r_{u,i}$ : rating dự đoán
- $\lambda$ : regularization
- tránh overfitting

---

### 3. Hybrid Collaborative Filtering

Hybrid Collaborative Filtering kết hợp:

- memory-based
- model-based
- content-based

Score tổng hợp:

$$
score
=
\alpha
\cdot CF
+
(1-\alpha)
\cdot Content
$$

Trong đó:

- CF : collaborative filtering score
- Content : content-based score
- $\alpha$ : trọng số

Mục tiêu:

- giảm cold start
- giảm sparsity
- tăng accuracy

---

### 4. Deep Learning Collaborative Filtering

Deep learning mở rộng matrix factorization:

$$
\hat r_{u,i}
=
f(U_u, V_i)
$$

Trong đó:

- f là neural network
- thay vì dot product tuyến tính

Ví dụ:
```
embedding(user) → dense → score embedding(item) → dense
```

---

### 5. Context-aware Collaborative Filtering

Collaborative filtering truyền thống:

$$
R_{u,i}
$$

Context-aware:

$$
R_{u,i,c}
$$

Trong đó:

- u : user
- i : item
- c : context (time, location...)

Ví dụ:

- sáng → nhạc nhẹ
- tối → nhạc EDM

Dữ liệu trở thành tensor 3 chiều.

---

### Thách thức của Collaborative Filtering

Mặc dù Collaborative Filtering là một trong những phương pháp phổ biến nhất trong hệ thống gợi ý, phương pháp này vẫn tồn tại nhiều thách thức khi áp dụng trên các tập dữ liệu lớn trong thực tế. Các vấn đề này chủ yếu xuất phát từ đặc điểm của dữ liệu tương tác người dùng – đối tượng, vốn thường rất lớn, thưa thớt và thay đổi liên tục theo thời gian. Những hạn chế này ảnh hưởng trực tiếp đến độ chính xác, khả năng mở rộng và tính ổn định của hệ thống.

---

### Dữ liệu thưa thớt (Data Sparsity)

Trong các hệ thống recommender thực tế, số lượng người dùng và đối tượng thường rất lớn, trong khi mỗi người dùng chỉ tương tác với một số lượng rất nhỏ các đối tượng. Điều này dẫn đến ma trận user–item:

$$
R_{u,i}
$$

trở nên cực kỳ thưa thớt. Ví dụ, trong một hệ thống có hàng triệu người dùng và hàng trăm nghìn sản phẩm, mỗi người dùng có thể chỉ mua hoặc đánh giá vài chục sản phẩm. Khi đó, phần lớn các phần tử trong ma trận sẽ là giá trị thiếu. Sự thưa thớt này làm cho việc tính toán độ tương đồng giữa người dùng hoặc giữa các đối tượng trở nên khó khăn, bởi vì số lượng phần tử được đánh giá chung là rất ít. Điều này làm giảm độ chính xác của các phương pháp dựa trên similarity, đồng thời làm tăng độ nhiễu trong quá trình dự đoán.

Một hệ quả quan trọng của dữ liệu thưa thớt là hệ thống khó học được sở thích thực sự của người dùng. Khi dữ liệu quá ít, các dự đoán dễ bị thiên lệch hoặc không ổn định. Ngoài ra, dữ liệu sparse cũng làm giảm hiệu quả của các thuật toán neighborhood-based vì việc tìm kiếm người dùng tương tự trở nên khó khăn hơn.

---

### Cold Start Problem

Vấn đề khởi đầu lạnh xảy ra khi hệ thống không có đủ dữ liệu để đưa ra dự đoán đáng tin cậy. Điều này xảy ra trong hai trường hợp chính: người dùng mới và đối tượng mới.

Trong trường hợp người dùng mới, hệ thống chưa có thông tin về lịch sử tương tác của người dùng đó. Vì Collaborative Filtering dựa trên hành vi trước đây để dự đoán sở thích, nên khi không có dữ liệu, hệ thống không thể xác định người dùng tương tự hoặc các item phù hợp. Do đó, các đề xuất ban đầu thường không chính xác hoặc mang tính ngẫu nhiên. Hệ thống chỉ có thể cải thiện khi người dùng bắt đầu tương tác với các item, ví dụ như rating, click hoặc purchase.

Tương tự, khi một item mới được thêm vào hệ thống, item đó chưa có rating hoặc interaction từ người dùng. Collaborative Filtering không thể xác định item đó giống với item nào khác. Do đó, item mới sẽ không được recommend cho đến khi có đủ người dùng tương tác với nó. Điều này tạo ra độ trễ trong việc khám phá các item mới và làm giảm khả năng đề xuất các sản phẩm mới.

Khác với Collaborative Filtering, các phương pháp content-based có thể giảm vấn đề này vì chúng sử dụng đặc trưng nội dung của item thay vì phụ thuộc hoàn toàn vào interaction.

---

### Khả năng mở rộng (Scalability)

Khi số lượng người dùng và đối tượng tăng lên, Collaborative Filtering gặp vấn đề nghiêm trọng về khả năng mở rộng. Giả sử có:

- $M$ người dùng
- $N$ đối tượng

Ma trận tương tác có kích thước:

$$
M \times N
$$

Các thuật toán memory-based thường cần tính similarity giữa các user hoặc item. Nếu tính similarity giữa user, độ phức tạp có thể là:

$$
O(M^2)
$$

Nếu tính similarity giữa item:

$$
O(N^2)
$$

Khi M và N rất lớn (hàng triệu), chi phí tính toán trở nên rất cao. Ngoài ra, nhiều hệ thống recommender cần phản hồi theo thời gian thực, do đó việc tính toán toàn bộ similarity là không khả thi. Điều này buộc hệ thống phải sử dụng các kỹ thuật tối ưu như approximate nearest neighbor, sampling hoặc candidate generation.

Khả năng mở rộng cũng liên quan đến việc cập nhật dữ liệu mới. Khi có user hoặc item mới, hệ thống có thể cần tính toán lại similarity, điều này làm tăng chi phí và độ trễ của hệ thống.

---

### Synonym Problem (Vấn đề từ đồng nghĩa)

Một vấn đề khác của Collaborative Filtering là các item giống nhau nhưng có tên hoặc biểu diễn khác nhau. Hệ thống không nhận ra rằng chúng có nội dung tương đương và xử lý chúng như các item độc lập. Ví dụ, hai item như "phim thiếu nhi" và "phim dành cho trẻ em" có thể biểu diễn cùng một loại nội dung nhưng được xem là khác nhau trong hệ thống. Điều này làm phân tán dữ liệu tương tác và giảm độ chính xác của recommendation.

Vấn đề này đặc biệt phổ biến trong hệ thống có dữ liệu văn bản hoặc metadata không chuẩn hóa. Một cách tiếp cận để giải quyết là sử dụng topic modeling hoặc embedding để nhóm các item có ý nghĩa tương tự vào cùng một không gian biểu diễn.

---

### Gray Sheep Problem

Gray sheep là những người dùng có hành vi không nhất quán với bất kỳ nhóm người dùng nào. Những người dùng này có sở thích đa dạng hoặc thay đổi liên tục, khiến việc tìm user tương tự trở nên khó khăn. Collaborative Filtering dựa vào similarity, nên khi user không giống ai, hệ thống không thể đưa ra đề xuất chính xác.

Ngoài ra, còn tồn tại black sheep, là những người dùng có sở thích hoàn toàn khác biệt so với phần còn lại. Trong trường hợp này, việc đề xuất gần như không thể thực hiện hiệu quả. Đây được xem là hạn chế tự nhiên của Collaborative Filtering.

---

### Shilling Attacks

Trong các hệ thống mở, người dùng có thể tự do đánh giá item. Điều này dẫn đến khả năng người dùng thao túng hệ thống bằng cách:

- rating cao cho sản phẩm của họ
- rating thấp cho sản phẩm đối thủ

Các hành vi này được gọi là shilling attacks. Nếu không có cơ chế phát hiện, hệ thống recommendation có thể bị lệch và đưa ra đề xuất sai. Đây là vấn đề quan trọng trong các hệ thống thương mại điện tử hoặc review online.

---

### Diversity và Long Tail Problem

Collaborative Filtering thường có xu hướng đề xuất các item phổ biến vì chúng có nhiều interaction hơn. Điều này tạo ra hiệu ứng "rich-get-richer", trong đó các item phổ biến ngày càng được đề xuất nhiều hơn. Ngược lại, các item ít phổ biến, dù có thể phù hợp hơn với người dùng, lại ít được đề xuất.

Hiện tượng này làm giảm tính đa dạng của recommendation và hạn chế khả năng khám phá item mới. Ngoài ra, nó cũng ảnh hưởng đến long-tail items, là các item ít phổ biến nhưng chiếm phần lớn catalogue. Nhiều nghiên cứu đã đề xuất các thuật toán để tăng diversity bằng cách đưa thêm các item mới, bất ngờ hoặc ít phổ biến vào danh sách gợi ý.
