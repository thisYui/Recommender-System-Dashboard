# Khái niệm

RS là đưa ra một hệ thống dự đoán mức độ phù hợp giữa người dùng (user) và đối tượng (item). Định nghĩa toán học thuần túy:

$$ f(user, item) -> score $$

TA lặp lại hành động trên đối với nhiều đối tượng (item) khác nhau và xắp xếp chúng theo thứ tự từ cao đến thấp và lấy $K$ điểm cao nhất gọi là $Top-K$
